   **A) Identify key trends and Weather.Condition patterns in aviation accidents to understand common causes.


```python
plt.figure(figsize=(6, 4))
weather_condition_counts.plot(kind='line', color='red')
plt.xlabel("Weather Condition")
plt.ylabel("Number of Accidents")
plt.title("Weather Conditions During Aviation Accidents")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

    **ANALYSIS for Weather Condition Trends

-Visual Meteorological Conditions (VMC) are weather conditions under which pilots have sufficient visibility to fly aircraft relying on visual references.Most of the Aircraft accidents occure during that weather conditions due to some factors like pilot errors
-IMC,Not Availlable,UNK weather conditions were not high which could be caused from poor visibilities.
    

**B) trends in accident occurrences can be observed from different phases of flight

```python
plt.figure(figsize=(10, 6))
broad_phase_percent.sort_index().plot(kind="bar", color=['grey', 'purple', 'green','yellow','red'], alpha=0.5)
plt.xlabel("broad_phase_of_flight")
plt.ylabel('broad_phase_counts ')
plt.title("Accidents by broad_phase_of_flight ")

plt.show()
```

**C) Evaluate high-risk regions to see if certain geographical areas have more accidents.

```python

location_counts = df["Location"].value_counts().head(30)

plt.figure(figsize=(12, 6))
plt.scatter(location_counts.index, location_counts.values, color="red", marker="o", s=100, alpha=0.5)
plt.xlabel("Location")
plt.ylabel("Number of Accidents")
plt.title("Top 10 High-Risk Regions for Aviation Accidents")
plt.xticks(rotation=45)

plt.show()
```

    **Analysis for certain geographical areas that have more accidents.
   
   The visualization  shows dots representing accident counts for each location, with the highest-risk locations.Anchorage AK is the most affected with cases of Aircraft accident due to volatile weather.other  locations have significantlyvlower accidents  counts but they need to be investigated to curb the factors.

**D-Identify how many aircrafts result in minor damage, substantial damage, or extensive damage in certain locations after the accident.


```python
damage_counts = df.groupby(['Location', 'Aircraft.damage']).size().unstack(fill_value=0)
top_damage_counts = damage_counts.sum(axis=1).nlargest(20).index
top_locations = damage_counts.loc[top_damage_counts]
top_locations.plot(kind='bar', stacked=True, figsize=(12, 6), colormap="viridis")

plt.title('Aircraft Damage by Location')
plt.xlabel('Location')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=75)
plt.legend(title='Damage Level')

plt.show()
```

    **Analysis for Aircraft damages Extent in diffrent locations
    
   Across most locations, Substantial Damage is the leading category.This suggests many accidents result in serious but not total aircraft loss. Identify if certain aircraft models or makes are more prone to severe damage to mitigate the accidents
