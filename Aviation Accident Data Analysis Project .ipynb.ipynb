######   **A) Identify key trends and Weather.Condition patterns in aviation accidents to understand common causes.


```python
plt.figure(figsize=(6, 4))
weather_condition_counts.plot(kind='line', color='red')
plt.xlabel("Weather Condition")
plt.ylabel("Number of Accidents")
plt.title("Weather Conditions During Aviation Accidents")
plt.xticks(rotation=45)
plt.grid()
plt.savefig("Aircraft Accidents by Weather Conditions ")  
plt.show()
```

######    **ANALYSIS for Weather Condition Trends

-Visual Meteorological Conditions (VMC) are weather conditions under which pilots have sufficient visibility to fly aircraft relying on visual references.IMC (Instrument Meteorological Conditions) accounted for fewer accidents.Most of the Aircraft accidents occure during that weather conditions due to some factors like pilot errors
-IMC,Not Availlable,UNK weather conditions were not high which could be caused from poor visibilities.
    

###### **B) trends in accident occurrences can be observed from different phases of flight

```python
plt.figure(figsize=(10, 6))
broad_phase_percent.sort_index().plot(kind="bar", color=['grey', 'purple', 'green','yellow','red'], alpha=0.5)
plt.xlabel("broad_phase_of_flight")
plt.ylabel('broad_phase_counts ')
plt.title("Accidents by broad_phase_of_flight ")
plt.savefig("Accidents by broad_phase_of_flight ")  
plt.show()
```

######    **ANAYSIS for broad_phase_of_flight
  
.Landing phase had the highest percentage of accidents (17.36%)
.Takeoff and cruise phases followed with 14.05% and 11.55% respectively
.Strengthening runway safety measures to minimize maneuvering accidents.
    

###### **C) Evaluate high-risk regions to see if certain geographical areas have more accidents.

```python

location_counts = df["Location"].value_counts().head(30)

plt.figure(figsize=(12, 6))
plt.scatter(location_counts.index, location_counts.values, color="red", marker="o", s=100, alpha=0.5)
plt.xlabel("Location")
plt.ylabel("Number of Accidents")
plt.title("Top 10 High-Risk Regions for A Accidents")
plt.xticks(rotation=45)
plt.savefig("Top 10 High-Risk Regions for Aircrafts Accidents")
plt.show()
```

######    **Analysis for certain geographical areas that have more accidents.
   
 Anchorage, AK had the highest number of accidents (434),Miami, FL (200) and Albuquerque, NM (196) followed affected with cases of Aircraft accident due to volatile weather.other  locations have significantlyvlower accidents  counts but they need to be investigated to curb the factors.

###### **D-Identify how many aircrafts result in minor damage, substantial damage, or extensive damage in certain locations after the accident.


```python
damage_counts = df.groupby(['Location', 'Aircraft.damage']).size().unstack(fill_value=0)
top_damage_counts = damage_counts.sum(axis=1).nlargest(20).index
top_locations = damage_counts.loc[top_damage_counts]
top_locations.plot(kind='bar', stacked=True, figsize=(12, 6), colormap="viridis")

plt.title('Aircraft Damage by Location')
plt.xlabel('Location')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=75)
plt.legend(title='Damage Level')
plt.savefig('Aircraft Damage by Location')
plt.show()
```

######    **Analysis for Aircraft damages Extent in diffrent locations
  
Substantial damage occurred in 72.17% of cases.Destroyed aircraft accounted for 20.95%.Minor damage was only 3.16%. Identify if certain aircraft models or makes are more prone to severe damage to mitigate the accidents
