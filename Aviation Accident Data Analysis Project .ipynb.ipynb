**Evaluate high-risk regions to see if certain geographical areas have more accidents.

```python

location_counts = df["Location"].value_counts().head(30)

plt.figure(figsize=(12, 6))
plt.scatter(location_counts.index, location_counts.values, color="red", marker="o", s=100, alpha=0.7)
plt.xlabel("Location")
plt.ylabel("Number of Accidents")
plt.title("Top 10 High-Risk Regions for Aviation Accidents")
plt.xticks(rotation=45)

plt.show()
```

**5-Identify how many aircrafts result in minor damage, substantial damage, or extensive damage in certain locations after the accident.


```python
plt.figure(figsize=(10, 6))
damage_counts.plot(kind='barh', color='green',alpha=0.4)
plt.xlabel('Damage Severity')
plt.ylabel('damage_counts')
plt.title('Aircraft Damage Severity in Accidents')
plt.xticks(rotation=0)
plt.show()
```

```python
damage_counts = df.groupby(['Location', 'Aircraft.damage']).size().unstack(fill_value=0)
top_damage_counts = damage_counts.sum(axis=1).nlargest(20).index
top_locations = damage_counts.loc[top_damage_counts]
top_locations.plot(kind='bar', stacked=True, figsize=(12, 6), colormap="viridis")

plt.title('Aircraft Damage by Location')
plt.xlabel('Location')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=75)
plt.legend(title='Damage Level')

plt.show()
```

    **Analysis for Aircraft damages Extent in diffrent locations
    
   Across most locations, Substantial Damage is the leading category.This suggests many accidents result in serious but not total aircraft loss. Identify if certain aircraft models or makes are more prone to severe damage to mitigate the accidents
